```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Comercios:
    def __init__(self):
        global df
        df = pd.read_csv('comercios.csv',index_col=0, parse_dates=True,thousands=r',', sep=';')
        print (df)
        
    def grafica(self,articulo):
        zona = df.groupby(['Articulo']).get_group((articulo))
        zona.groupby(['Zona','Comercio'], sort=False)['Precio'].max().unstack(fill_value=0).plot(kind='bar',stacked=False)
        plt.title(articulo)
        plt.show()
```

```python
dato = Comercios()
```

```python
dato.grafica('Agua')
dato.grafica('Macarrones')
```
