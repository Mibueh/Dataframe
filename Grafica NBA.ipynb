```python
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
class Nba:
    def excel(self):
        global nba
        global group
        excel_file = 'NBA.xlsx'
        nba = pd.read_excel(excel_file, index_col='Nombre')
        group = nba.sort_values(['Posicion'], ascending=False)
        group.head(50)
        
    def posicion(self):
        global clave
        fig=plt.figure(figsize=(30,10))
        plt.subplot2grid((2,4),(0,0))
        clave = nba['Posicion']
        clave.value_counts().plot(kind='bar', alpha=0.5)
        plt.title('Posiciones m√°s jugadas')
        plt.show()
        
    def pivot(self):
        fig=plt.figure(figsize=(30,10))
        plt.subplot2grid((2,4),(0,0))
        pivot = nba.groupby(['Posicion']).get_group(('Pivot'))
        rebote = pivot.sort_values(['Rebotes'], ascending=False)
        rebote.head(10).plot(kind="bar", alpha=0.5)
        plt.title('Pivot')
        plt.show()
        
    def jugador(self,zona):
        player = nba.groupby(['Posicion']).get_group((zona))
        asistencia = player.sort_values(['Asistencia'], ascending=False)
        asistencia.plot(kind="bar")
        plt.title(zona)
        plt.figure(figsize=(20,10))
        plt.show()
        
    def alero(self):
        alero = nba.groupby(['Posicion']).get_group(('Alero'))
        triple = pivot.sort_values(['Triples'], ascending=False)
        triple.head(10).plot(kind="bar")
        plt.title('Alero')
        plt.show()
        
    def escolta(self):
        escolta = nba.groupby(['Posicion']).get_group(('Escolta'))
        puntos = escolta.sort_values(['Puntos'], ascending=False)
        puntos.head(10).plot(kind="bar")
        plt.title('Escolta')
        plt.show()
```
